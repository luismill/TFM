```python
import pandas as pd
import numpy as np
```

```python
# Importación del driver de Cassandra
from cassandra.cluster import Cluster
# Configuración de la conexión
cluster = Cluster([''])
session = cluster.connect('tecnologia_energia')
```

```python
path = r''
excel = pd.read_excel(path, 'Direct values')
excel.head(2)
```

```python
excel.shape
```

```python
excel.isna().sum()
```

```python
excel.drop(['KKS ENDING', 'PI&D', 'SHEET', 'LOCATION'], inplace=True, axis=1)
```

```python
excel[excel.isnull().any(axis=1)]
```

```python
kks = excel['KKS COMPLETE'][250].split('::')[1]
kks
```

```python
excel.columns
```

```python
excel.head()
```

```python
kks = []
i=0
for row in excel.itertuples():
    if not isinstance(row._4,str):
        lectura = row.KKS
    else:
        lectura = row._4.split('::')[1].replace('/','_')
    kks.append(lectura)
    if isinstance(row.HIERACHY,str):
        i += 1

print(i)
```

```python
kks_cassandra = []
rows = session.execute('SELECT * FROM kks')
for r in rows:
    kks_cassandra.append(r.kks)
```

```python
print('KKS en Cassandra: ', len(kks_cassandra))
print('KKS en el excel: ', len(kks))
print('KKS en ambos: ', len(set(kks) & set(kks_cassandra)))
```

```python
query = session.prepare("SELECT * from kks where kks=? and plant='shagaya'")
rows = session.execute(query, [kks,])
for r in rows:
    print(r)
```
